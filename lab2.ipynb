## Thực hành: 
- Thông kế độ dài trung bình của mỗi từ = Tổng ký tự / Tổng số từ
- Độ dài trung bình của mỗi câu = Tổng số từ / Tổng số câu
- Số lần xuất hiện trung bình của một từ = Tổng số từ đơn / Tổng số từ

<!>
- Lưu ý: gutenberg.raw() Không trả về list mà trả về tổng số lượng kí tự có chức cả dấu cách
- gutenberg.sents() trả về list các câu. Mỗi câu là list các từ

```python
from nltk.corpus import gutenberg
print('Độ dài TB từ'.ljust(20), 'Độ dài TB câu'.ljust(20), 'Số lần xuất hiên'.ljust(20), 'Văn bản')
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid)) # Tính số ký tự
    num_words = len(gutenberg.words(fileid)) # Tính số từ
    num_sents = len(gutenberg.sents(fileid)) # Tính số câu
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)])) # Tính số từ không trùng nhau
    print(str(int(num_chars/num_words)).ljust(20), str(int(num_words/num_sents)).ljust(20), str(int(num_words/num_vocab)).ljust(20), fileid)
```

# Brown Corpus
Corpus điện tử đầu tiên bằng tiếng Anh (1961), chứa văn bản từ 500 nguồn và các nguồn thì được chia mục.

```python
from nltk.corpus import brown
brown.categories()
```

Lấy ra các từ thuộc chuyên mục có tên là "news"

```python
brown.words(categories='news')
```

Lấy ra các câu, từ những chuyên mục khác nhau:

```python
brown.sents(categories=['news', 'editorial', 'reviews'])
```

Thống kê về cách sử dụng các động từ khuyết thiếu (Model verb) 

```python
cfd = nltk.ConditionalFreqDist((genre, word) for genre in brown.categories() for word in brown.words(categories=genre))
```

```python
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
print(cfd.tabulate(conditions=genres, samples=modals))
```

### Tạo corpus

```python
import os
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
corpusdir = 'newcorpus/' # Directory of corpus.
newcorpus = PlaintextCorpusReader(corpusdir, '.*')
```

```python
# Access each file in the corpus.
for fileid in newcorpus.fileids():
    print(fileid)   # Tên file
    with newcorpus.open(fileid) as f:  # Mở file
        print(f.read().strip()) # Đọc file
```

```python
# Access paragraphs in a coprus file. ( sử dụng nltk.tokenize.sent_tokenize() và nltk.tokenize.word_tokenize() )
print( newcorpus.paras() ) 
```
